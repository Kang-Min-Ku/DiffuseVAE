```python
# Replot in case of failure to display the graph
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of logvar values
logvar = np.linspace(-5, 5, 500)

# Calculate weights for the three options
epsilon = 1e-6
var = np.exp(logvar)  # Variance

# Option 1: Inverse scaling
W_option1 = 1 / (1 + np.exp(var))

# Option 2: Sigmoid transformation
W_option2 = 1 / (1 + np.exp(logvar))

# Option 3: Soft clipping
W_option3 = 1 / (var + 1)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(var, W_option1, label="Option 1: Inverse Scaling")
plt.plot(var, W_option2, label="Option 2: Sigmoid Transformation")
plt.plot(var, W_option3, label="Option 3: Soft Clipping")
plt.xscale("log")
plt.xlabel("Variance (log scale)")
plt.ylabel("Weight (W)")
plt.title("Variance vs. Weight (W) for Adaptive Weight Options")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
